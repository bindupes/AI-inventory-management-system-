```python
import pandas as pd

# Step 1: Read the CSV file
df = pd.read_csv('INVENTORY6.csv')

# Convert PRICE column to float to handle numerical operations
df['PRICE'] = pd.to_numeric(df['PRICE'], errors='coerce')

# Step 2: Define a function to adjust the price based on the category and manufactured price
def adjust_price(Category, PRICE):
    if Category == 'Electronics':
        if PRICE > 4000:
            return PRICE - 1000
        else:
            return PRICE - 500
    elif Category == 'Cosmetics':
        if PRICE > 300:
            return PRICE - 50
        else:
            return PRICE - 30
    elif Category == 'Food & breverages':
        if PRICE > 300:
            return PRICE - 50
        else:
            return PRICE - 30
    elif Category == 'Groceries':
        if PRICE > 300:
            return PRICE - 40
        elif 200 <= PRICE <= 300:
            return PRICE - 30
        else:
            return PRICE - 20
    elif Category == 'Clothing':
        if 550 < PRICE < 1000:
            return PRICE - 60
        elif PRICE > 1000:
            return PRICE - 70
        else:
            return PRICE - 30
    elif Category == 'Stationary':  # Note the capitalization should match your CSV
        if PRICE > 250:
            return PRICE - 50
        else:
            return PRICE - 30
    elif Category == 'Furniture':
        if PRICE > 10000:
            return PRICE - 2500
        elif 5000 <= PRICE <= 10000:
            return PRICE - 1500
        else:
            return PRICE - 500
    else:
        return PRICE  # If category is unknown, return the price unchanged

# Step 3: Define a function to calculate the final price for the given quantity
def calculate_final_price(PRODUCT_NAME, Quantity):
    # Find the product in the CSV, assuming PRODUCT_NAME is unique
    product = df[df['PRODUCT_NAME'] == PRODUCT_NAME].iloc[0]
    category = product['Category']
    manufactured_price = product['PRICE']
    
    # Adjust the price based on the category and manufactured price
    adjusted_price = adjust_price(category, manufactured_price)
    
    # Multiply the adjusted price by the quantity (adjust price only once)
    final_price = adjusted_price * Quantity
    
    return final_price

# Example usage
PRODUCT_NAME = 'TRESemme Keratin Smooth Shampoo'  # Example product name
Quantity = 2  # Example quantity

final_price = calculate_final_price(PRODUCT_NAME, Quantity)
print(f"The final price for {Quantity} {PRODUCT_NAME}(s) is: {final_price}")
```
